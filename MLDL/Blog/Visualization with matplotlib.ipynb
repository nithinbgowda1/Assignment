Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(0,9)
x
```

```python
y = np.arange(11,20)
y
```

```python
plt.plot(x,y)            #To plot the graph we use plot function
plt.xlabel('X-Axis')     # This gives label to the xaxis
plt.ylabel('Y Axis')    # This gives label to y-axis
plt.title('This is simple graph plotting') # This gives title to the graph
plt.savefig('test1.jpg')
plt.show()
```

```python
y = x*x
```

```python
plt.plot(x,y,'g*', linestyle = "dashed", linewidth = 4, markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
```

Subplot: To display multiple graph in the canvas

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r*',linestyle = "dashed", linewidth = 4, markersize = 10)
plt.subplot(2,2,3)
plt.plot(x,y,'bo',linestyle = "solid", linewidth = 4, markersize = 10)
plt.subplot(2,2,2)
plt.plot(x,y,linestyle = "solid", linewidth = 4, markersize = 10)
```

Generating sin waves or cosine waves

```python
np.pi
```

```python
np.arange(0,5*np.pi,0.1)
```

```python
x = np.arange(0,5*np.pi,0.1)
ysin = np.sin(x)
ycos = np.cos(x)
plt.subplot(2,1,1)
plt.plot(x,ysin)
plt.title('Sine Graph')
plt.subplot(2,1,2)
plt.title('Cosine Graph\n')
plt.plot(x,ycos)
```

box plot To calculate the percentile

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data,vert = True, patch_artist=False)
plt.show()
```

pie chart

```python
labels = ['C++','Ruby','Perl','Java']
size = [40,50,80,20]
explode = [0,0,0,0.5]
plt.pie(size, labels = labels, explode = explode, autopct = '%1.1f%%', shadow=False)
plt.axis('equal')
plt.show()
```
